```python
import pandas as pd   #Importing everything
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

### Getting the weekly *case* statistics for Vermont. 
I slice down the general information to the correct dates and state, then calculate using .mean, .median, and .mode.

```python
df = pd.read_csv('covid_confirmed_usafacts.csv') #imports the csv

dfsliced = df.loc[:, '2020-06-01':'2021-01-03']  #Slice data to only pertinant dates

dfweeksliced = dfsliced[dfsliced.columns[::7]] #Weekly dates
dflastday = dfsliced[dfsliced.columns[-1:]] #Last day as asked in instructions

covid2020 = pd.concat([df.iloc[:, :4], dfweeksliced], axis=1) #Combines first columns
covidcalc = pd.concat([covid2020, dflastday], axis = 1) #Adds last column to sliced data

STATE = 'Vermont'  #First state
STATE_ABBR = 'VT'
```

## Calculating the mean, median, and mode.

```python
vt2020 = covidcalc[covidcalc['State'] == STATE_ABBR]  #Reduce to only Vermont
vtmean = vt2020.loc[:, '2020-06-01':'2021-01-03'].mean(numeric_only=True)
vtmedian = vt2020.loc[:, '2020-06-01':'2021-01-03'].median(numeric_only=True)   #Mean, median, and mode calculations.
vtmode = vt2020.loc[:, '2020-06-01':'2021-01-03'].mode(numeric_only=True, dropna=True, axis = 1)
```

### Vermont Mean Cases Graph Generation

```python
plt.figure(figsize=(8,6))
plt.ylabel("Vermont Mean Cases")   #Graph detailing
plt.xlabel("Date")

vtmean.plot()
```

### Vermont Median Cases Graph Generation

```python
plt.figure(figsize=(8,6))
plt.ylabel("Vermont Median Cases")   #Graph detailing
plt.xlabel("Date")

vtmedian.plot()
```

#### Merging all the case information into a single dataframe

```python
state_cases_int = state_cases_merged.select_dtypes(include='int')
state_cases_normalized = (state_cases_int - state_cases_int.min()) / (state_cases_int.max() - state_cases_int.min())  #normalizer acquired

scn_num = state_cases_normalized.drop(['countyFIPS','StateFIPS'], axis = 1)

plt.figure(figsize=(8,6))
plt.ylabel("State Cases")   #Graph detailing
plt.xlabel("Date")

scn_num.plot()
```
